# Life Expectancy analysis of the 6 Nations


With the analysis of the scatter plot below, it is evident that the majority of these nations apart from Zimbabwe have maintained a high rate of life expectancy ranging from 70 to 80 years old with Germany having the best trend with an increase from the 78 year old mark which was recorded in 2000 to 82 in 2014. Another interesting finding from these graphs is that although Zimbabwe is the most undeveloped country in terms of life expectancy readings, they are showing an upward trajectory, showing improvement every year, indicating that this country is heading towards the right direction in their efforts to improve the life expectancy of their people.

# Average Life Expectancy Graph 

The analysis of the average life expectancy of these nations, the majority of these countries have an With the analysis of the average life expectancy of these nations, the country with the best average life expectancy is Germany with an average life expectancy rate being 79.7. The country with the worst average life expectancy is Zimbabwe with figures being significantly low in comparison to the other 3 countries, close to 50.

```python
countries = ['United States of America','China','Chile','Mexico', 'Zimbabwe','Germany']
countries_graph  = ['USA', 'China', 'Chile', 'Mexico', 'Zimbabwe', 'Germany']
average_life_expectancy = []
```

```python
def life_expectancy_append(countries):
    for country in countries: 
        average_life_expectancy.append(float(country_dataframe[country_dataframe['Country'] == country]['Life Expectancy'].mean()))

life_expectancy_append(countries)
print(average_life_expectancy)

plt.bar(countries_graph, average_life_expectancy, color = 'Green')
plt.title('Life Expectancy averages for 6 countries')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.show()
```

# Individual Life Expectancy Graphs for each country from 2000- 2015

The readings from these individual graphs indicate that the countries that are keeping a strong level of life expectancy readings and projecting an increasing upward trajectory are China,USA and Germany. Although Chile and Mexico are showing life expectancy readings that are improved, they don't show the same level of constant upward trajectory like China, USA and Germany. 

```python
def scatterplot_maker(country):
    sns.scatterplot(data = country_dataframe[country_dataframe.Country == country], x = 'Year', y = 'Life Expectancy')
    plt.title('Life Expectancy for ' + country)
    plt.show()

for country in countries: 
    scatterplot_maker(country)
```
